```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dropping the unneccessary column

```python
df = df.drop("Unnamed: 0", axis= 1)
```

```python
df.head()
```

# gender distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean', 'ReadingScore':"mean", "WritingScore":"mean"})
```

```python
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.show()
#plt.save('heatmap.png')
```

```python
gm= df.groupby('ParentMaritalStatus').agg({'MathScore':'mean', 'ReadingScore':"mean", "WritingScore":"mean"})
```

```python
print(gm)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gm, annot=True)
plt.show()
```

# Distribution in EthnicGroup

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ['groupA', 'groupB', 'groupC', 'groupD', 'groupE']
lst = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.pie(lst, labels = l, autopct='%1.2f%%')
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "EthnicGroup")
ax.bar_label(ax.containers[0])
plt.show()
```
